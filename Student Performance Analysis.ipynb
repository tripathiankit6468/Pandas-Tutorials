```python
import pandas as pd
```

```python
df = pd.read_csv('students.csv')
```

```python
print(df)
```

```python
print(df.head())
print(df.info())
print(df.describe())
print(df.shape)
```

```python
print(df.isnull().sum())
```

```python
df['Total Marks'] = df['Maths'] + df['Science'] + df['English']
print(df)
```

```python
df['Percentage'] = df['Total Marks'] / 3
print(df)
```

```python
top_student = df.loc[df['Percentage'].idxmax()]
print(top_student)
```

```python
high_performers = df[df["Percentage"] > 80]
print(high_performers)
```

```python
sorted_stud_perc = df.sort_values(by = "Percentage" , ascending = False)
print(sorted_stud_perc)
```

```python
gender_avg = df.groupby('Gender')['Percentage'].mean()
print(gender_avg)
```

```python
class_avg = df.groupby('Class')['Percentage'].mean()
print(class_avg)
```

```python
def grade(p):
    if p >= 90:
        return "A+"
    elif p >= 80:
        return "A"
    elif p >= 70:
        return "B"
    elif p >= 60:
        return "C"
    else:
        return "D"

df["Grade"] = df["Percentage"].apply(grade)

print(df)
```

```python
print("Total Students :", len(df))
print("Average Percentage :", round(df["Percentage"].mean(), 2))
print("Highest Percentage :", round(df["Percentage"].max(), 2))
print("Lowest Percentage :", round(df["Percentage"].min(), 2))
```

```python
df.to_csv("student_report.csv", index=False)

print("Report Saved Successfully")
```
